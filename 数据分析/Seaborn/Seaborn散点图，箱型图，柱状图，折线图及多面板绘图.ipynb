## 分类散点图

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style = 'whitegrid',color_codes = True)
np.random.seed(sum(map(ord,'categorical')))

tips = sns.load_dataset('tips')

sns.stripplot(x = 'day',y = 'total_bill',data = tips)
plt.show()
```

散射点通常会重叠。这使得我们很难看到完整的数据分布。一个简单的解决方法是使用一些随机的“抖动”来调整位置。

```python
sns.stripplot(x = 'day',y = 'total_bill',jitter = True,data = tips)
plt.show()
```

使用`swarmplot()`,避免散点重叠带分布的散点图。

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = tips)
plt.show()
```

使用`hue`参数添加一个分类变量。

```python
sns.swarmplot(x = 'day',y = 'total_bill',hue = 'sex',data = tips)
plt.show()
```

分类变量放在y轴上。

```python
sns.swarmplot(x = 'total_bill',y = 'day',hue = 'time',data = tips)
plt.show()
```

## 箱形图

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style = 'whitegrid',color_codes = True)
np.random.seed(sum(map(ord,'categorical')))

tips = sns.load_dataset('tips')

sns.boxplot(x = 'day',y = 'total_bill',hue = 'time',data = tips)
plt.show()
```

自定义分组。

```python
tips['weekend'] = tips['day'].isin(['Sat','Sun'])
sns.boxplot(x = 'day',y = 'total_bill',hue = 'weekend',data = tips)
plt.show()
```

箱线图+有分布趋势的散点图。

```python
#图形组合也就是两条绘图语句一起运行就可以了，相当于图形覆盖了
sns.boxplot(x = 'day',y = 'total_bill',data = tips)
sns.swarmplot(x = 'day',y = 'total_bill',data = tips,color = '0.25')
plt.show()
```

## 柱状图

### 分类柱状图

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style = 'darkgrid',color_codes = True)
titanic = sns.load_dataset('titanic')

sns.barplot(x = 'sex',y = 'survived',hue = 'class',data = titanic,capsize = .05)
plt.show()
```

### 变量计数，绘制柱状图

```python
sns.countplot(x = 'deck',data = titanic,palette = 'Greens_d')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y="deck", data=titanic, color="c")
plt.show()
```

### 因子变量分类统计

```python
sns.countplot(y = 'deck',hue = 'class',data = titanic,palette = 'Greens_d')
plt.show()
```

```python
sns.countplot(x="class", hue="who", data=titanic)
plt.show()
```

### 水平堆砌条形图

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
#设置子图的大小
f, ax = plt.subplots(figsize=(6, 15))
#导入数据集car crash dataset
crashes = sns.load_dataset("car_crashes").sort_values("total", ascending=False)

sns.set_color_codes("pastel")
sns.barplot(x="total", y="abbrev", data=crashes,label="Total", color="b")
sns.set_color_codes("muted")
sns.barplot(x="alcohol", y="abbrev", data=crashes,label="Alcohol-involved", color="b")

#设置图例和坐标信息
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 24), ylabel="",xlabel="Automobile collisions per billion miles")
sns.despine(left=True, bottom=True)
plt.show()
```

## 折线图

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
titanic = sns.load_dataset("titanic")

sns.pointplot(x="deck", y="survived", hue="sex", data=titanic,palette={"male": "g", "female": "m"},markers=["^", "o"], linestyles=["-", "--"],capsize=0.1)
plt.show()
```

```python
sns.pointplot(x="class", y="survived", hue="sex", data=titanic,palette={"male": "g", "female": "b"},markers=["^", "o"], linestyles=["-", "--"],capsize=0.1)
plt.show()
```

## 多面板绘图

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
tips = sns.load_dataset("tips")

sns.factorplot(x="time", y="total_bill", hue="smoker",col="day", data=tips, kind="box", size=4, aspect=.5,legend=False)
plt.legend(loc = 'best',title ='smoker')
plt.show()
```
